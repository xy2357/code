```python
users['country'] = users['country'].str.strip().str.title()
users
```

```python
users['age'] = users['age'].astype(float)
users
```

```python
users['signup_date'] = pd.to_datetime(users['signup_date'])
users
```

```python
orders['quantity'] = orders['quantity'].astype(float)
orders['discount'] = orders['discount'].astype(float)
```

```python
orders['order_time'] = pd.to_datetime(orders['order_time'])
orders
```

```python
orders = orders.drop_duplicates(
    subset = ['user_id', 'product_id', 'quantity', 'order_time', 'status', 'discount'],
    keep = 'first'    
)
```

```python
orders['discount'] = orders['discount'].fillna(0)
orders
```

```python
filt = users['country'] == 'China'
users[filt]
```

```python
filt = (users['age'] > 25) & (users['gender'] == 'female')
users[filt]
```

```python
filt = (orders['status'] == 'paid')
orders[filt]
```

```python
# filt = orders['order_time'].dt.month == 2
# orders[filt]
filt = (orders['order_time'] >= '2024-02-01') & (orders['order_time'] < '2024-03-01')
orders[filt]
```

```python
orders['month'] = orders['order_time'].dt.strftime('%Y') + '-' + orders['order_time'].dt.strftime('%m')
orders
```

```python
orders['hour'] = orders['order_time'].dt.strftime('%H')
orders
```

```python
orders['is_weekend'] = orders['order_time'].dt.dayofweek > 5
orders
```

```python
orders['final_amount'] = products.loc[orders['product_id'], 'price'] * orders['quantity'] - orders['discount']
```
